```python
import requests
import pandas as pd
from sqlalchemy import create_engine
!pip install pymysql


# 1. Configuración de conexión a MySQL
config_mysql = {
    'usuario': 'root',
    'contraseña': '1234',
    'host': 'localhost',
    'puerto': '3307',
    'base_de_datos': 'platzi_api'
}
```

```python
import requests
import pandas as pd

def obtener_datos_productos():
    url = "https://api.escuelajs.co/api/v1/products"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        # Normalizar los datos anidados
        df_api_normalize = pd.json_normalize(data)
        print("✅ Datos normalizados correctamente.")
        return df_api_normalize

    except requests.exceptions.RequestException as e:
        print(f"❌ Error al obtener productos: {e}")
        return None

# Llamar la función y guardar el DataFrame
df_api_normalize = obtener_datos_productos()

# Verificar que no sea None antes de usarlo
if df_api_normalize is not None:
    df_api_normalize.info()
else:
    print("No se pudieron cargar los datos.")
```

```python
def limpiar_datos(df):
    # Eliminar valores nulos
    df_clean = df.dropna()

    # Convertir lista de imágenes a cadena separada por comas
    df_clean['images'] = df_clean['images'].apply(lambda imgs: ', '.join(imgs) if isinstance(imgs, list) else str(imgs))

    # Eliminar duplicados ignorando 'images'
    df_clean = df_clean.drop_duplicates(subset=[col for col in df_clean.columns if col != 'images'])

    # Asegurar tipo numérico para el precio
    df_clean['price'] = pd.to_numeric(df_clean['price'], errors='coerce').fillna(0).astype(int)

    return df_clean
```

```python
# 4. Exportar a MySQL
def exportar_a_mysql(df, nombre_tabla, config):
    try:
        cadena_conexion = f"mysql+pymysql://{config['usuario']}:{config['contraseña']}@{config['host']}:{config['puerto']}/{config['base_de_datos']}"
        engine = create_engine(cadena_conexion)

        with engine.connect() as conn:
            print("✅ Conexión a MySQL exitosa")

        df.to_sql(nombre_tabla, engine, if_exists='replace', index=False)
        print(f"\n✅ Datos exportados a la tabla '{nombre_tabla}'")
        print(f"🧾 Total de productos exportados: {len(df)}")

    except Exception as e:
        print(f"\n❌ Error al exportar a MySQL: {e}")
        print("Verifica:")
        print(f"- Usuario: {config['usuario']}")
        print(f"- Contraseña: {'*' * len(config['contraseña'])}")
        print(f"- Host: {config['host']}")
        print(f"- Puerto: {config['puerto']}")
        print(f"- Base de datos: {config['base_de_datos']}")
        print("\n¿Está corriendo el servidor MySQL? ¿Tienes los privilegios necesarios?")

# === EJECUCIÓN PRINCIPAL ===
df_api_normalize = obtener_datos_productos()

if df_api_normalize is not None:
    df_limpio = limpiar_datos(df_api_normalize)

    print("\n📌 Vista previa de los datos procesados:")
    print(df_limpio.head())

    exportar_a_mysql(df_limpio, 'productos_api', config_mysql)
else:
    print("No se pudieron obtener productos desde la API.")
```
